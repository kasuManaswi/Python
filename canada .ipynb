```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('canada .csv')
```

```python
df
```

```python
sns.regplot(x='year',y='per capita income (US$)',data=df)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
brain=LinearRegression()
```

```python
#train the model
a=df.drop('per capita income (US$)',axis=1)
```

```python
a
```

```python
b=df.drop('year',axis=1)
```

```python
b
```

```python
#train the model
brain.fit(a,b)
```

```python
#to measure accuracy
brain.score(a,b)
```

```python
brain.predict([[2025]])
```
